## Performance and Analysis

####  To obtain depth,we need to  take disparity output and use the following formula:
\begin{align}
\mathbf{depth} & = \frac{baseline *  focal}{disparity*width}
\end{align}


● The baseline is the distance between the two cameras

● The focal is the focal length

● Disparity is the output of the network in form of disparities.

● Width is the image width in pixels.
